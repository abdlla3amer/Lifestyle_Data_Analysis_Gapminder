*FREE DATA FROM WORLD BANK VIA GAPMINDER.ORG, CC-BY LICENSE* <br>

Gapminder has collected a lot of information about how people live their lives in different countries, tracked across the years, and on a number of different indicators.

#### Indicators choosen for this invistigation

##### 1- Children per woman (total fertility rate)
- Description: Total fertility rate. The number of children that would be born to each woman with precailing age-specific fertility rates.
- Source url: https://www.gapminder.org/data/documentation/gd008/
- csv file downloaded from: gapminder.org/data/ .. Choose individual indicators .. Babies per woman
- file name: children_per_woman_total_fertility.csv
- Dataset alias: df_tfr
- Dataset shape: 202 rows (countries) 302 columns (years from 1800 to 2100)


##### 2- Child Mortality Rate
- Description: Death of children under five years of age per 1,000 live births.
- Source url: https://www.gapminder.org/data/documentation/gd005/
- csv file downloaded from: gapminder.org/data/ .. Choose individual indicators .. Child mortality.
- file name: child_mortality_0_5_year_olds_dying_per_1000_born.csv
- Dataset alias: df_cmr
- Dataset shape: 197 rows (countries) 302 columns (years from 1800 to 2100)


##### 3- Employment rate % (females aged 15+)
- Description: Percentage of female population, age group 15+, that has been employed during the given year.
- Source url: https://ilostat.ilo.org/
- csv file downloaded from: gapminder.org/data/ .. Choose individual indicators .. Work .. Employment rate .. Female aged 15+
- file name: females_aged_15plus_employment_rate_percent.csv
- Dataset alias: df_erg
- Dataset shape: 189 rows (countries) 30 columns (years from 1991 to 2019)

- Missing data can be filled with data from the previous year (forward fill method) which in this case makes more sense than filling data with the mean.

```python
# converting strings above into lists
# spliting strings around '\n' and excloding the last element which is '\n' too.
africa_countries = africa_countries.split('\n')[:-1]
asia_countries = asia_countries.split('\n')[:-1]
europe_countries = europe_countries.split('\n')[:-1]
north_america_countries = north_america_countries.split('\n')[:-1]
south_america_countries = south_america_countries.split('\n')[:-1]
australia_countries = australia_countries.split('\n')[:-1]
```

```python
print('Africa Countries', africa_countries)
```

```python
# defining a new function that checks the country and returns its continent
def country_continent(country):
    if country in africa_countries:
        return 'Africa'
    elif country in asia_countries:
        return 'Asia'
    elif country in europe_countries:
        return 'Europe'
    elif country in north_america_countries:
        return 'North America'
    elif country in south_america_countries:
        return 'South America'
    elif country in australia_countries:
        return 'Australia'
    else:
        return 'Unknown'
```

```python
# iterating over the datasets
for df in [df_tfr, df_cmr, df_erg]:
    #iterating over 'country' column which is the index
    for country in df.index:
        df['country'] = df.index
        df['continent_'] = df['country'].apply(country_continent)
# inserting the 'continent' column as the first column after the index and dropping 'continent_'  and 'country' column
for df in [df_tfr, df_cmr, df_erg]:
    df.insert(0, 'continent', df['continent_'])
    df.drop(df[['continent_', 'country']], axis=1, inplace=True)
```

```python
# checking for 'Unknown' continents and trying to fill them manually
print(df_tfr.query('continent == "Unknown"').shape[0])
print(df_cmr.query('continent == "Unknown"').shape[0])
print(df_erg.query('continent == "Unknown"').shape[0])
```

### Research Question 2  (Are there any trends in Continents Total Fertility Rates over the years?)

- Africa had the highest mean Total Fertility Rates over the years.
- Europe had the lowest mean Total Fertility Rates over the years
- Except for Europe, the curves were quite steady untill the 1950s.
- Except for Europe, the mean Total Fertility Rate was significantly ascending between 1950 and 1960 and then the curves started descending fastly unitl 2020, afirca had the same approach with some delay.
- Europe had a descending curve from the beginning (1800) until 2000, but the descendece between 1900 and 2000 was alot faster than the descendence between 1800 and 1900.
- Europe is the only Continent that had an ascending curve (inceasing in mean total fertility rates) between 2000 and 2020.
